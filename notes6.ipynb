```python
housing_dataframe = pd.read_csv('HousingData.csv')
housing_dataframe.head(6) 
```

```python
housing_dataframe.info() 
```

```python
housing_dataframe.describe() 
```

```python
# Null Values, 
# Popular choices to deal w null values 
# 1. Eliminate the rows
# 2. Replace with significant value, ie median or the mean. 
# 3. replace with the most likely value 
```

```python
# find the rows and columns in the dataset with null values 
housing_dataframe.isnull().any() 
```

```python
# Replacing Null Values 
housing_dataframe['AGE'] = housing_dataframe['AGE'].fillna(housing_dataframe.mean()) 
housing_dataframe['CHAS'] = housing_dataframe['CHAS'].fillna(0) 
housing_dataframe = housing_dataframe.fillna(housing_dataframe.median()) 
housing_dataframe
```

```python
housing_dataframe.info() 
```

```python
# Visual Analysis, 
# matplotlib Library 
import matplotlib.pyplot as plt 
%matplotlib inline 
import seaborn as sns 
sns.set() 
```

```python
plt.hist(housing_dataframe['MEDV'])
plt.show() 
```

```python
plt.hist(housing_dataframe['MEDV'])
plt.title('Median Boston Housing Prices')
plt.xlabel('1980 Median Value in Thousands')
plt.ylabel('Count')
plt.show() 
```

```python
# Define Functions 
def my_hist(column, title, xlab, ylab):
    plt.figure(figsize=(10, 6))
    plt.hist(column) 
    plt.title(title, fontsize=15)
    plt.xlabel(xlab) 
    plt.ylabel(ylab) 
    plt.savefig(title, dpi=300) 
    plt.show() 
```

```python
my_hist(housing_dataframe['RM'], 'Average Number of Rooms in Bostom Households', 'Average Number of Rooms', 'Count')
```

```python
# scatter plots
x = housing_dataframe['RM']
y = housing_dataframe['MEDV']
plt.scatter(x, y) 
plt.show() 
```

```python
# Correlations, statistical measure between -1 and +1 that indicates how closely two variables are related. 
# correlations of -1 or +1 means that varibles re dependent and fall perfectly in a striaght line 
# correlation of 0 indicates that an increase in one variable gives no information whatsoever abt the other variable 
# correlation of 0.75 means strong relationship 
# correlation of 0.25 means weak relationship 
```

```python
housing_dataframe.corr() 
```

```python
corr = housing_dataframe.corr() 
plt.figure(figsize=(8, 6))
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values, cmap="Blues", linewidths=1.25, alpha=0.8)
plt.show() 
```

```python
# Regression, a line that tries to fit the values of a scatter plot as closely as possible 
plt.figure(figsize=(10, 7))
sns.regplot(x, y) 
plt.show() 
```

```python
# StatsmODEL Regression Output 
import statsmodels.api as sm 
X = sm.add_constant(x) 
model = sm.OLS(y, X) 
est = model.fit() 
print(est.summary()) 
```
